```python
!pip install llama-cpp-python

from llama_cpp import Llama

llm = Llama(
    model_path   = "/content/qwen2.5-0.5b-instruct-q4_k_m.gguf",
    n_ctx        = 2048,
    n_gpu_layers = 35,   # use GPU on Colab for this test
    verbose      = False,
)

result = llm.create_chat_completion(
    messages = [
        {"role": "system",  "content": "You are a precise HR data reporter. Answer in one sentence."},
        {"role": "user",    "content": "Facts:\n  name = Dr. A Sharma\n  attendance_pct = 88.5\n  days_present = 212\n  working_days = 239\n\nQuestion: What is the attendance percentage?\nAnswer:"},
    ],
    max_tokens  = 80,
    temperature = 0.15,
)

print(result["choices"][0]["message"]["content"])
```
